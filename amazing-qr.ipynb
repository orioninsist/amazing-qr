### 🖼️ Arka Plan Görseli / GIF Yükleme (Opsiyonel)
Eğer QR kodunuzun arkasına bir resim veya GIF eklemek istiyorsanız, aşağıdaki hücreyi çalıştırarak bilgisayarınızdan yükleyin.
**Not:** Yüklediğiniz dosya aşağıdaki ayarlarda otomatik olarak kullanılacaktır.

```python
# @title 📤 Dosya Yükle { display-mode: "form" }
from google.colab import files
import os

print("📁 Arka plan için bir resim veya GIF seçin (Opsiyonel).")
print("İptal ederseniz standart QR kod oluşturulacaktır.\n")

uploaded = files.upload()
uploaded_file_name = ""

if uploaded:
    uploaded_file_name = list(uploaded.keys())[0]
    print(f"\n✅ Yüklendi: {uploaded_file_name}")
    print("💡 Artık aşağıdaki 'QR Kod Oluşturucu' hücresinde bu dosyayı kullanabilirsiniz.")
else:
    print("\nℹ️ Herhangi bir dosya yüklenmedi. Standart QR oluşturulacak.")
```

### 🎨 QR Kod Oluşturucu
Parametreleri yan taraftaki formdan doldurun ve hücreyi çalıştırın.

```python
# @title ⚙️ Parametre Ayarları { display-mode: "form" }

Words = "https://github.com/orioninsist" # @param {type:"string"}
Version = 1 # @param {type:"slider", min:1, max:40, step:1}
Level = "H" # @param ["L", "M", "Q", "H"]
Use_Background = True # @param {type:"boolean"}
Background_Path = "uploaded" # @param {type:"string"}
Colorized = True # @param {type:"boolean"}
Contrast = 1.0 # @param {type:"number"}
Brightness = 1.0 # @param {type:"number"}
Output_Name = "amazing_qr.png" # @param {type:"string"}
Output_Dir = "/content/output" # @param {type:"string"}

import os

final_picture = ""

if Use_Background:
    if Background_Path.lower() == "uploaded":
        if 'uploaded_file_name' in globals() and uploaded_file_name:
            if os.path.exists(uploaded_file_name):
                final_picture = uploaded_file_name
            else:
                print(f"⚠️ Yüklenen dosya ({uploaded_file_name}) bulunamadı! Standart QR oluşturulacak.")
        else:
            print("⚠️ Henüz bir dosya yüklemediniz! Lütfen üstteki hücreden dosya yükleyin veya manuel yol girin.")
    else:
        if os.path.exists(Background_Path):
            final_picture = Background_Path
        else:
            print(f"⚠️ Belirtilen dosya ({Background_Path}) bulunamadı! Standart QR oluşturulacak.")

if final_picture:
    print(f"📝 Kullanılan Arka Plan: {final_picture}")
    # Eğer GIF yüklendiyse çıktı adını otomatik .gif yap
    if final_picture.lower().endswith('.gif') and Output_Name.lower().endswith('.png'):
        Output_Name = Output_Name.rsplit('.', 1)[0] + '.gif'
        print(f"📝 Çıktı formatı GIF olarak güncellendi: {Output_Name}")

# Komutu oluştur
cmd = f'amzqr "{Words}" -v {Version} -l {Level} -n "{Output_Name}" -d "{Output_Dir}" -con {Contrast} -bri {Brightness}'

if final_picture:
    cmd += f' -p "{final_picture}"'
    if Colorized:
        cmd += ' -c'

print(f"🏃 Çalıştırılıyor: {cmd}")
!{cmd}

# Sonucu göster
from IPython.display import Image, display, HTML
import os
import base64

output_path = os.path.join(Output_Dir, Output_Name)
if os.path.exists(output_path):
    print("\n✨ İşte QR Kodunuz:")
    if Output_Name.lower().endswith('.gif'):
        with open(output_path, "rb") as f:
            data = base64.b64encode(f.read()).decode("utf-8")
        display(HTML(f'<img src="data:image/gif;base64,{data}" width="400">'))
    else:
        display(Image(filename=output_path))
else:
    print("\n❌ Dosya oluşturulamadı. Lütfen parametreleri kontrol edin.")
```

### 📁 Çıktıları İndirme
Oluşturulan dosyayı aşağıdaki kodu çalıştırarak indirebilirsiniz.

```python
# @title 💾 İndir { display-mode: "form" }
from google.colab import files
import os

try:
    output_path = os.path.join(Output_Dir, Output_Name)
    if os.path.exists(output_path):
        files.download(output_path)
    else:
        print("İndirilecek dosya bulunamadı.")
except NameError:
    print("Lütfen önce QR kod oluşturma hücresini çalıştırın.")
```
