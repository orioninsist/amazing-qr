### 🎨 QR Kod Oluşturucu
Parametreleri yan taraftaki formdan doldurun ve hücreyi çalıştırın.

```python
# @title ⚙️ Parametre Ayarları { display-mode: "form" }

Words = "https://github.com/orioninsist" # @param {type:"string"}
Version = 1 # @param {type:"slider", min:1, max:40, step:1}
Level = "H" # @param ["L", "M", "Q", "H"]
Picture = "" # @param {type:"string"}
Colorized = True # @param {type:"boolean"}
Contrast = 1.0 # @param {type:"number"}
Brightness = 1.0 # @param {type:"number"}
Output_Name = "amazing_qr.png" # @param {type:"string"}
Output_Dir = "/content/output" # @param {type:"string"}

import os

# Komutu oluştur
cmd = f'amzqr "{Words}" -v {Version} -l {Level} -n "{Output_Name}" -d "{Output_Dir}" -con {Contrast} -bri {Brightness}'

if Picture:
    cmd += f' -p "{Picture}"'

if Colorized:
    cmd += ' -c'

print(f"🏃 Çalıştırılıyor: {cmd}")
!{cmd}

# Sonucu göster
from IPython.display import Image, display
output_path = os.path.join(Output_Dir, Output_Name)
if os.path.exists(output_path):
    print("\n✨ İşte QR Kodunuz:")
    display(Image(filename=output_path))
else:
    print("\n❌ Dosya oluşturulamadı. Lütfen parametreleri kontrol edin.")
```

### 📁 Çıktıları İndirme
Oluşturulan dosyaları sol taraftaki dosya menüsünden (`output` klasörü) veya aşağıdaki kodu çalıştırarak indirebilirsiniz.

```python
from google.colab import files
import os

output_path = os.path.join(Output_Dir, Output_Name)
if os.path.exists(output_path):
    files.download(output_path)
else:
    print("İndirilecek dosya bulunamadı.")
```
